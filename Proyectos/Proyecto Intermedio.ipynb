```python
import random
import matplotlib.pyplot as plt

# Configuración del gráfico
fig, ax = plt.subplots()
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)
ax.set_aspect('equal')
ax.add_artist(plt.Circle((0, 0), 1, fill=False, color='black'))
plt.title("Aproximación de π por método de Monte Carlo")

# Generación de puntos aleatorios y conteo
num_puntos_totales = 1000
num_puntos_circulo = 0
for i in range(num_puntos_totales):
    x = random.uniform(-1, 1)
    y = random.uniform(-1, 1)
    distancia = x**2 + y**2
    if distancia <= 1:
        num_puntos_circulo += 1
        plt.plot(x, y, 'ro')
    else:
        plt.plot(x, y, 'bo')

# Estimación de π y visualización del gráfico
estimate_pi = 4 * num_puntos_circulo / num_puntos_totales
pi=3.14159
print("Estimación de π:", estimacion_pi)
print("Número de puntos dentro del círculo:", num_puntos_circulo)
print("Número total de puntos:", num_puntos_totales)
plt.show()
```

```python
Para estimar cuántos puntos necesitamos generar para obtener una precisión de ±0.01, podemos hacer lo siguiente:

Sabemos que la fórmula para la estimación de π utilizando el método de Monte Carlo es:
π ≈ 4 * (puntos dentro del círculo) / (puntos totales)
La precisión que queremos alcanzar es de ±0.01, lo que significa que queremos que nuestra estimación de π se encuentre dentro de un rango de ±0.01 del valor real de π.
Podemos usar la desigualdad de Chebyshev para encontrar una estimación de la cantidad de puntos que necesitamos generar para lograr esta precisión. 
La desigualdad de Chebyshev establece que para cualquier distribución de probabilidad, al menos 1 - 1/k^2 de los valores estarán dentro de k desviaciones estándar de la media. En este caso, podemos aplicar la desigualdad de Chebyshev a nuestra estimación de π y establecer lo siguiente:
|π - π_est| ≤ 0.01
|π - 4 * (puntos dentro del círculo) / (puntos totales)| ≤ 0.01
|4 * (puntos dentro del círculo) / (puntos totales) - π| ≤ 0.01
|(puntos dentro del círculo) / (puntos totales) - π/4| ≤ 0.0025
Aquí, π_est es nuestra estimación de π, y estamos buscando una precisión de ±0.01, lo que significa que el valor real de π debe estar dentro de un rango de ±0.01 de nuestra estimación.
Podemos establecer k = 0.05 (0.0025 / 0.05^2) para que al menos el 95% de las estimaciones caigan dentro del rango deseado.
De la fórmula de la desigualdad de Chebyshev, podemos despejar el número mínimo de puntos necesarios para lograr la precisión deseada:
n ≥ (1/k^2) * (π/4 * (1 - π/4)) / 0.01^2
n ≥ (1/0.05^2) * (π/4 * (1 - π/4)) / 0.01^2
n ≥ 7812.5
Esto significa que necesitamos generar al menos 7813 puntos para lograr una precisión de ±0.01 con una probabilidad del 95%.
En la práctica, es posible que necesitemos generar más puntos para lograr una precisión precisa. Por lo tanto, se recomienda ejecutar el algoritmo varias veces y tomar el promedio de las estimaciones de π para obtener una estimación más precisa.
```

```python
import matplotlib.pyplot as plt

num_points = [100, 500, 1000, 5000, 10000, 50000, 100000]
errors = []
for n in num_points:
    pi_approx = estimate_pi
    errors.append(abs(pi - pi_approx))
```

```python
plt.plot(num_points, errors)
plt.xscale('log')
plt.xlabel('Número de puntos')
plt.ylabel('Error')
plt.title('Error de la estimación de π')
plt.show()

```
